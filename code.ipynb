```python
#dealing with range values
# Function to process values
def process_value(value):
    if '-' in str(value):
        # If it's a range, split and calculate the average
        start, end = map(float, str(value).split('-'))
        return (start + end) / 2
    else:
        # If it's not a range, convert to float
        return float(value)

# Apply the function to the column
df['acceleration_0-100km/h'] = df['acceleration_0-100km/h'].apply(process_value)

print(df['acceleration_0-100km/h'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import re


# Function to extract and convert weight-to-power ratio values to a consistent numeric format
def extract_and_convert_weight_to_power_ratio(value):
    try:
        # Use regular expressions to find numeric values (e.g., '3.3' or '299.1')
        numeric_values = re.findall(r'\d+\.\d+', value)
        if numeric_values:
            # Extract the first numeric value found and convert it to a float
            numeric_value = float(numeric_values[0])
            return numeric_value
        else:
            return None  # Return None for values that cannot be parsed
    except:
        return None

# Apply the function to extract and convert weight-to-power ratio values
df['weight-to-power_ratio_numeric'] = df['weight-to-power_ratio'].apply(extract_and_convert_weight_to_power_ratio)

# Create a plot with the extracted numeric values
plt.figure(figsize=(8, 6))
plt.scatter(df['weight-to-power_ratio_numeric'], df['acceleration_0-100km/h'], marker='o', linestyle='-')
plt.title('Acceleration (0-100km/h) vs. Weight-to-Power Ratio')
plt.xlabel('Weight-to-Power Ratio (Numeric)')
plt.ylabel('Acceleration (0-100km/h)')
plt.grid(True)

# Show the plot
plt.show()
```

```python
def extract_and_convert_torque(value):
    try:
        # Use regular expressions to find numeric values (e.g., '570')
        numeric_values = re.findall(r'\d+\.\d+|\d+', value)
        if numeric_values:
            # Extract the first numeric value found and convert it to an integer
            numeric_value = int(numeric_values[0])
            return numeric_value
        else:
            return None  # Return None for values that cannot be parsed
    except:
        return None

# Apply the function to extract and convert torque values
df['torque_numeric'] = df['torque'].apply(extract_and_convert_torque)

# Create a scatter plot with the extracted numeric torque values
plt.figure(figsize=(8, 6))
plt.scatter(df['acceleration_0-100km/h'], df['torque_numeric'], marker='o')
plt.title('Scatter Plot of Acceleration (0-100km/h) vs. Torque')
plt.xlabel('Acceleration (0-100km/h)')
plt.ylabel('Torque (Nm)')
plt.grid(True)

# Show the scatter plot
plt.show()
```

```python
import re
from sklearn.impute import SimpleImputer


# Define a custom function to handle non-numeric values
def handle_non_numeric(value):
    # Check if the value is numeric (float or int)
    if pd.notna(value) and isinstance(value, (int, float)):
        return value  # Numeric value, no change needed
    elif pd.notna(value) and re.match(r'^\d+-\d+$', str(value)):
        # Handle values like "14-18" by taking the average
        range_values = [int(x) for x in str(value).split('-')]
        return sum(range_values) / len(range_values)
    else:
        return None  # Replace non-numeric values with None

```

```python
# Apply the custom function to the column
df['fuel_consumption_combined'] = df['fuel_consumption_combined'].apply(handle_non_numeric)

# Create a SimpleImputer instance with strategy='mean' to perform mean imputation
imputer = SimpleImputer(strategy='mean')

# Apply mean imputation to the column
df['fuel_consumption_combined'] = imputer.fit_transform(df[['fuel_consumption_combined']])

```

```python


import re
from sklearn.impute import SimpleImputer


# Define a custom function to handle non-numeric values and calculate the average for ranges
def handle_non_numeric_and_ranges(column):
    def inner(value):
        # Check if the value is numeric (float or int)
        if pd.notna(value) and isinstance(value, (int, float)):
            return value  # Numeric value, no change needed
        elif pd.notna(value) and re.match(r'^\d+-\d+$', str(value)):
            # Handle values like "1372-1398" by taking the average
            range_values = [int(x) for x in str(value).split('-')]
            return sum(range_values) / len(range_values)
        else:
            return None  # Replace non-numeric values with None

    return column.apply(inner)

# List of columns to process
columns_to_process = ['front_track', 'rear_track', 'drag_coefficient']

# Apply the custom function to each column in the list
for column in columns_to_process:
    df[column] = handle_non_numeric_and_ranges(df[column])

# Create a SimpleImputer instance with strategy='mean' to perform mean imputation for remaining non-numeric values
imputer = SimpleImputer(strategy='mean')

# Apply mean imputation to the columns
df[columns_to_process] = imputer.fit_transform(df[columns_to_process])

# Print the DataFrame with values processed and missing values imputed with the mean
print(df)

```

```python
def remove_ranges_and_average(column):
    # Regular expression pattern to match a range like "1235-1278"
    pattern = r'(\d+)-(\d+)'

    def replace_with_avg(match):
        start = int(match.group(1))
        end = int(match.group(2))
        avg = (start + end) / 2
        return str(avg)

    # Use regular expressions to find and replace ranges in the column
    column = column.apply(lambda x: re.sub(pattern, replace_with_avg, str(x)))
    return column

# Apply the custom function to the columns where you want to remove ranges
columns_to_process = ['kerb_weight' , 'max_load' , 'width' , 'height']
for column in columns_to_process:
    df[column] = remove_ranges_and_average(df[column])

# Convert the columns to numeric if needed
df[columns_to_process] = df[columns_to_process].apply(pd.to_numeric, errors='coerce')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#label encoder


# Separate categorical and numerical columns
categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
numerical_columns = df.select_dtypes(include=['number']).columns.tolist()

encoded_df = pd.get_dummies(df, columns=categorical_columns)

print(encoded_df)
```

```python
#target variable is maximum_speed

X = encoded_df.drop('maximum_speed',axis=1)
y = encoded_df['maximum_speed']
```

Removing features with low variance - Variance Threshold

```python
from sklearn.feature_selection import VarianceThreshold

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=.30)


vth = VarianceThreshold(threshold=0)  # as deafult threshold=0
vth.fit(X_train)
X_train_vth = X_train.iloc[:, vth.get_support()]

df1 = pd.DataFrame( {'Feature': X_train.columns,'Variance': vth.variances_,}).sort_values('Variance', ascending=True)
df1
```

```python
#removing features w low variance

from sklearn.feature_selection import VarianceThreshold

# Create a VarianceThreshold instance with a threshold (e.g., 0 for zero variance)
selector = VarianceThreshold(threshold=0)

# Fit the selector on your data
selector.fit(X_train)

# Get the mask of non-constant features
mask = selector.get_support()

# Apply the mask to your data to keep only non-constant features
X_train = X_train.loc[:, mask]
X_test = X_test.loc[:, mask]
```

```python
#Select KBest

import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest, f_regression

# Create a SelectKBest instance with f_regression as the scoring function
Kbest_regression = SelectKBest(score_func=f_regression, k='all')   
Kbest_regression.fit(X_train, y_train)

# Get the feature scores and names
feature_scores = Kbest_regression.scores_
feature_names = X_train.columns

# Create a dictionary to store feature scores and names
feature_score_dict = dict(zip(feature_names, feature_scores))

# Sort the features by score in descending order
sorted_features = sorted(feature_score_dict.items(), key=lambda x: x[1], reverse=True)

# Select the top N features for plotting (e.g., top 10)
top_features = sorted_features[:30]

# Extract feature names and scores for the top N features
top_feature_names, top_feature_scores = zip(*top_features)

# Create a bar plot to visualize the top N feature scores
plt.bar(top_feature_names, top_feature_scores)
plt.xticks(rotation=90)
plt.rcParams["figure.figsize"] = (8, 6)
plt.title('Top 30 Feature Scores for Maximum Speed Prediction')
plt.xlabel('Features')
plt.ylabel('Scores')
plt.show()
```

```python
# Sort the features by score in descending order
data = sorted(list(zip(X_train.columns, Kbest_regression.scores_)), key=lambda x: x[1], reverse=True)

# Select the top 30 features
top_data = data[:30]

# Separate the top data into two lists
top_features = [item[0] for item in top_data]
top_scores = [item[1] for item in top_data]

# Create a horizontal bar plot for the top 30 features
plt.figure(figsize=(10, 8))
plt.barh(top_features, top_scores)
plt.xlabel('Scores')
plt.ylabel('Features')
plt.title('Top 30 Feature Importance Scores')
plt.gca().invert_yaxis()  # Reverse the order to show the most important features at the top
plt.show()
```

```python
# Generic code

from sklearn.metrics import (
    silhouette_score, calinski_harabasz_score, davies_bouldin_score,
    normalized_mutual_info_score, adjusted_rand_score,
    adjusted_mutual_info_score, v_measure_score,
    completeness_score, homogeneity_score
)

# Define a function to evaluate clustering metrics for a given model
def evaluate_clustering_metrics(model, data, scaled_data):
    labels = model.fit_predict(scaled_data)
    evaluation_metricsx = {
        'Silhouette Score': silhouette_score(scaled_data, labels),
        'Calinski-Harabasz Score': calinski_harabasz_score(scaled_data, labels),
        'Davies-Bouldin Score': davies_bouldin_score(scaled_data, labels),
        'Normalized Mutual Info': normalized_mutual_info_score(encoded_df['speed_category_Very High'].values, labels),
        'Adjusted Rand Index': adjusted_rand_score(encoded_df['speed_category_Very High'].values, labels),
        'Adjusted Mutual Info': adjusted_mutual_info_score(encoded_df['speed_category_Very High'].values, labels),
        'V-Measure': v_measure_score(encoded_df['speed_category_Very High'].values, labels),
        'Completeness Score': completeness_score(encoded_df['speed_category_Very High'].values, labels),
        'Homogeneity Score': homogeneity_score(encoded_df['speed_category_Very High'].values, labels),
    }
    return evaluation_metricsx

# Initialize a list to store clustering models
clustering_models = [
    KMeans(n_clusters=2),
    KMeans(n_clusters=3),
    AgglomerativeClustering(n_clusters=2),
    AgglomerativeClustering(n_clusters=3),
    SpectralClustering(n_clusters=2),
    SpectralClustering(n_clusters=3)
]

lx =[]
# Loop through all clustering models and evaluate metrics
for model in clustering_models:
    d = []
    print(f"Model: {model.__class__.__name__}")
    metrics = evaluate_clustering_metrics(model, data, scaled_data)
    for metric_name, score in metrics.items():
        print(f"{metric_name}: {score:.4f}")
        d.append(score)
    lx.append(d)
    print()
    
```
